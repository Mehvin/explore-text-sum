```python
# This program cleans useless information off CNN & Dailymail datasets
# These useless information are consistent across most (if not all) articles

import os

cnn_path = 'cnn/stories'
dm_path = 'dailymail/stories'

cnn_file_count = 0
cnn_itr = 1
dm_file_count = 0
dm_itr = 1

# Cleaning CNN files

# (CNN) ...
# Washington (CNN) ...

for filename in os.listdir(cnn_path):
    cnn_file_count = cnn_file_count + 1
    
    # Counter to check on progress
    if (cnn_file_count == 1000):
        print('Finished ' + str(1000 * cnn_itr) + ' CNN articles')
        cnn_file_count = 0
        cnn_itr = cnn_itr + 1
    
    with open(cnn_path + '/' + filename, 'r') as f:
        lines = f.readlines()

    if '(CNN)' in lines[0]:
        count = 0
        for char in lines[0]:
            count = count + 1
            if char == '-':
                break
                
        lines[0] = lines[0][count + 2:]

        with open(cnn_path + '/' + filename, 'w') as f:
            f.truncate()
            f.writelines(lines)
            

print('CNN Datasets Cleaned!')


# Cleaning Dailymail files

# By
# Olivia Williams

# PUBLISHED:

# 07:13 EST, 30 August 2013


# UPDATED:

# 08:38 EST, 22 August 2013

for filename in os.listdir(dm_path):
    dm_file_count = dm_file_count + 1
        
    # Counter to check on progress
    if (dm_file_count == 1000):
        print('Finished ' + str(1000 * dm_itr) + ' Dailymail articles')
        dm_file_count = 0
        dm_itr = dm_itr + 1
    
    with open(dm_path + '/' + filename, 'r') as f:
        lines = f.readlines()
    
    if 'By' in lines[0]:
        lines = lines[3:]

    published_line_no = 0
    for line in lines:
        if 'PUBLISHED:' in line:
            for x in range(1,10):
                try:
                    lines[published_line_no+x]
                    if lines[published_line_no+x] != '\n':
                        lines = lines[published_line_no+x+2:]
                        break
                except IndexError:
                    break

        published_line_no = published_line_no + 1
                    
    updated_line_no = 0
    for line in lines:
        if 'UPDATED:' in line:
            for x in range(1,10):
                try:
                    if lines[updated_line_no+x] != '\n':
                        lines = lines[updated_line_no+x+2:]
                        break
                except IndexError:
                    break
        
        updated_line_no = updated_line_no + 1
        
    with open(dm_path + '/' + filename, 'w') as f:
            f.truncate()
            f.writelines(lines)
        
print('Dailymail Datasets Cleaned!')
```
