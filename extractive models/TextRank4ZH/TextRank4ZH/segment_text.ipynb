```python
# Logging Module
import datetime
import logging
    
logger = logging.getLogger()
    
def setup_file_logger(log_file):
    hdlr = logging.FileHandler(log_file)
    formatter = logging.Formatter('%(message)s')
    hdlr.setFormatter(formatter)
    logger.addHandler(hdlr) 
    logger.setLevel(logging.INFO)
    
def log(message):
    #outputs to Jupyter console
    #print('{} {}'.format(datetime.datetime.now(), message))
    #outputs to file
    logger.info(message)
    
setup_file_logger('failed_to_run.log')
```

```python
#This program is done to segment Chinese text to prep results for ROUGE evaluation

part1_reference = "/home/s10166858/kawaijoe/Extractive Methods/TextRank4ZH/results_part_1/reference/"
part2_reference = "/home/s10166858/kawaijoe/Extractive Methods/TextRank4ZH/results_part_2/reference/"

part1_system = "/home/s10166858/kawaijoe/Extractive Methods/TextRank4ZH/results_part_1/system/"
part2_system = "/home/s10166858/kawaijoe/Extractive Methods/TextRank4ZH/results_part_2/system/"
            
import os
import sys
import codecs
import re
import jieba #Chinese segmentation tool
```

```python
# part1_reference
count = 0

for file in sorted(os.listdir(part1_reference), key=lambda name: int(name[7:-15])):
    try:
        count = count + 1
        f = codecs.open(part1_reference + file, 'r+', 'utf-8')
        lines = codecs.open(part1_reference + file, 'r+', 'utf-8').readlines()
        oldtext = codecs.open(part1_reference + file, 'r+', 'utf-8').read()
        
        #Jieba segmenting
        seg_list = jieba.cut(oldtext, cut_all=False) # Accurate mode - suitable for this case
        newtext = " ".join(seg_list)
        
        del lines[0]
        f.write(newtext)
        f.close()
        
    except:
        count = count + 1
        log("Part1_reference")
        log(file)
        log("Unexpected error:" + str(sys.exc_info()[0]))
        log("\n")
        continue
        
    if count % 1000 == 0:
        print(str(count) + " files ran")
```
