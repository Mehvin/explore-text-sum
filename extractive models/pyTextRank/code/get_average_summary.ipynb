```python
# This program is to find the average no of words & no of sentences
# for the 'gold standard' under reference folder
import os
reference_folder_path = "/home/kawaijoe/Extractive Methods/pytextrank/reference"
```

```python
# Get average number of words
no_of_words = 1
file_count = 0
itr = 1

for file in os.listdir(reference_folder_path):
    file_count = file_count + 1
    
    if file_count == 1000:
        print("Went through " + str(itr * file_count) + " files")
        itr = itr + 1
        file_count = 0
    
    with open(reference_folder_path + '/' + file, 'r') as f:
        for line in f:
            for letter in line:
                if letter == " ": # After a space == 1 word 
                    no_of_words = no_of_words + 1

print("Total no of words: " + str(no_of_words))
#23424017 - Letters
#4246636 - Words

print("Average no of words: " + str(no_of_words/93626))
#250.18709546493494 - Letters
#45.35744344519685 - Words
```

```python
# Get average number of sentenece
no_of_sentences = 0
file_count = 0
itr = 1

for file in os.listdir(reference_folder_path):
    file_count = file_count + 1
    
    if file_count == 1000:
        print("Went through " + str(itr * file_count) + " files")
        itr = itr + 1
        file_count = 0
        
    with open(reference_folder_path + '/' + file, 'r') as f:
        for line in f:
            no_of_sentences = no_of_sentences + 1
                
print("Total no of sentences: " + str(no_of_sentences))
#360420

print("Average no of sentences: " + str(no_of_sentences/93626))
#3.8495717001687564
```
